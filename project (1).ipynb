```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('datafile1.csv',skiprows=20)
```

```python
df2=pd.read_csv('datafile2.csv',skiprows=17)
```

```python
#merging two csvs
df=df=pd.merge(df1,df2,how='left',on=('LAT','LON','YEAR','MO','DY'))
df.head()
```

```python
df['MO']=df.MO.astype(str)
df['YEAR']=df.YEAR.astype(str)
df['DY']=df.DY.astype(str)

df['date']=df['YEAR'].str.cat(df['MO'],sep='/')
df['Date']=df['date'].str.cat(df['DY'],sep='/')
df.head()
```

```python
#removing lat and lon columns
df.drop(columns=['YEAR','MO','DY','LAT','LON','QV2M','date','WS50M_RANGE', 'WS50M_MIN','WS50M_MAX', 'WS50M','T2MDEW', 'T2MWET',
       'WS10M_MIN', 'WS10M_MAX','WS10M'],axis=1,inplace=True)
df.head()
```

```python
df.set_index(['Date'],inplace=True)
df.head()
```

```python
#scale the data
data_train=df.loc[:'2017/12/31',:][[  'PRECTOT', 'RH2M', 'PS', 'T2M_RANGE','TS', 'T2M_MAX', 'T2M_MIN', 'T2M','WS10M_RANGE']]
data_test=df.loc['2018/1/1':'2019/12/31',:][[  'PRECTOT', 'RH2M', 'PS', 'T2M_RANGE','TS' ,'T2M_MAX', 'T2M_MIN', 'T2M','WS10M_RANGE']]

from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
data_train_scaled=scalar.fit_transform(data_train)
data_test_scaled=scalar.transform(data_test)
```

```python
#create training and testing dataset
#splitting data into X and Y for training  
x_train,y_train=[],[]
#for j in range(9):
for i in range(len(data_train_scaled)):
  x_train.append(data_train_scaled[i-1])
  y_train.append(data_train_scaled[i])


#splitting data into X and Y for testing
x_test=[]
y_test=[]
#for j in range(9):
for i in range(len(data_test_scaled)):
  x_test.append(data_test_scaled[i-1])
  y_test.append(data_test_scaled[i])
```

```python
pd.DataFrame(x_train)
```

```python
pd.DataFrame(y_train)
```

```python
x_train,y_train=np.array(x_train),np.array(y_train)
x_train.shape,y_train.shape
```

```python
x_test=np.array(x_test)
y_test=np.array(y_test)
x_test.shape,y_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import Adam
from keras.models import load_model
```

```python
model = Sequential()
model.add(Dense(20, input_dim=9, activation='sigmoid'))
model.add(Dense(20, activation='sigmoid'))
#model.add(Dense(20, activation='sigmoid'))
model.add(Dense(9))
#model.add(Activation('sigmoid'))

model.summary()


model.compile(optimizer=Adam(learning_rate=0.1),loss='mse', metrics=[['accuracy'], ['mse']])
```

```python
history=model.fit(x_train,y_train,epochs=150,steps_per_epoch=300,batch_size=5,shuffle=False,validation_data=(x_test,y_test))
```

```python
# save model and architecture to single file
model.save("model.h5")
print("Saved model to disk")


```

```python
model = load_model('model.h5')
model.summary()
```

```python
model.compile(optimizer='adam',loss='mse', metrics=['accuracy'])
```

```python
train_predict=model.predict(x_train)
test_predict=model.predict(x_test)
```

```python
train_predict=scalar.inverse_transform(train_predict)
test_predict=scalar.inverse_transform(test_predict)
y_test=scalar.inverse_transform(y_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, test_predict)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,test_predict)
```

```python
scores = model.evaluate(x_test,y_test,verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
